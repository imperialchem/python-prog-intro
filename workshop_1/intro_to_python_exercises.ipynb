# Workshop 1 exercises

In these exercises you will write functions that perform a particular task. People often document their code by including strings inside functions that describe what the function does and what the arguments are. This string is called a *docstring* as is included directly below the function signature, it has no effect on the processing the function performs.

For the exercises we are using this string to specify how we want the function to behave. It shows examples of the function being used along with what output we expect to be produced. 

That sounds more complicated that it is so here's an example to get the idea.

```python
def identity(var):
    r'''
    >>> identity('a')
    'a'
    >>> identity(123)
    123
    >>> identity([1, 2, 3])
    [1, 2, 3]
    '''
```

Your job would be to add content to the identity function (below the string) such that calling this function with 'a' as an argument would mean the function returns 'a', likewise calling the function with 123 as an argument would return 123. 

The solution is to return the variable we are passing in:

```python
def identity(var):
    r'''
    >>> identity('a')
    'a'
    >>> identity(123)
    123
    >>> identity( [1, 2, 3] )
    [1, 2, 3]
    '''
    return var
```

NB when we solve the task by completing the function **we leave the docstring in place**.

Lets check it works the way we expect it to:

    identity('a')

    identity(123)

    identity([1, 2, 3])
